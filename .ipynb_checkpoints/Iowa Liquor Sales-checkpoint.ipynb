# Build your models

Using scikit-learn or statsmodels, build the necessary models for your scenario. Evaluate model fit.

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
                      



#define the FEATURES you will use - the variables that determine the outcome
X = df[["Bottles Sold"]]

#define the TARGET - the variable that you are trying to predict
y = df["Sale (Dollars)"]

#MODEL BUILDING

#define the type of model you will use - in this case, a linear regression
#store it as a variable so that it becomes an easy object to use
lm = linear_model.LinearRegression()

#fit the model on your data - use the data points to develop coefficients, etc. in the specific formula
#this of it as training the model to "use X to guess y, based on what we know to be true"
model = lm.fit(X, y)

#create a list of predictions using the regression (line) we created
#this will say that for a given point in X (defined by the features you chose), what would we expect y to be?
predictions = model.predict(X)

# VISUALIZE/PLOT THE MODEL

plt.scatter(predictions, y, s=30, c='r', marker='+', zorder=10)
plt.xlabel("Predicted Values")
plt.ylabel("Actual Price")
plt.show()

# SCORE THE MODEL 

#evaluate how well the model's predictions matched the actual values of y
print "MSE:", mean_squared_error(y, predictions)
import numpy as np


```

```python
#RMSE
np.sqrt(mean_squared_error(y, predictions)) #216.21142249869081
model.score(X,y) # R^2 0.68136076061510931
lm.intercept_ #-1.0834100912850033
lm.coef_ # array([ 13.15693634])

# "R-squared" (R^2) is proportion of variability in y that can be explained using x
```

```python
Interpreting Model Coefficients
How do we interpret the TV coefficient (β1)?

A unit increase in Bottles Sold is associated 13.15126217 "unit" in sales. Or more clearly. 
A "unit" increase in TV ad spending is associated with a 0.047537 "unit" increase in Sales.
Or more clearly. An additional bottle sold is associated with an increase in Sales 
Or more clearly: An additional $1,000 spent on TV ads is associated with an increase in sales of 47.537 widgets.

Using the Model for Prediction
Let's say that there was a new market where the TV advertising spend was $50,000. What would we predict for the Sales in that market?
```
